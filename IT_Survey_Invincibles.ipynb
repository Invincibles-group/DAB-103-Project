### Reading the CSV file

```python
df = pd.read_csv('Europe_IT_Salary_Survery.csv')
#df = pd.read_csv('3_years_Survey_excel_forgery.csv')
```

# Data Cleaning and Preparation

```python
#Checking Null Values Count
df.isnull().sum()
```

# Renaming Columns

```python
df2=df.rename(columns={"Year":"Year_of_Response",
                       "Your main technology / programming language":"Primary_Skillset",
                       "Yearly brutto salary (without bonus and stocks) in EUR":"Salary_Without_Incentives",
                       "Yearly bonus + stocks in EUR":"Annual_Incentives",
					   "Position ":"Position",
					   "Total years of experience":"Years_of_Experience",
					   "Main language at work":"Communication Language",
                       "Ð¡ontract duration":"Contract_Duration",
                       "Have you lost your job due to the coronavirus outbreak?":"Unemployed_CovidOutbreak"})
```

# Dropping Columns

```python
df2=df2.drop(['Years of experience in Germany',
         'Other technologies/programming languages you use often',
         'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country',
         'Annual bonus+stocks one year ago. Only answer if staying in same country',
         'Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week',
         'Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'],axis=1)
```

```python
df2.isnull().sum()
```

# Dropping Null Values

```python
df3= df2.dropna(subset=['Age','Gender','City','Position','Years_of_Experience','Seniority level'])
```

```python
df3.isnull().sum()
```

# Dropping Null Values

```python
df4= df3.dropna(subset=['Salary_Without_Incentives','Communication Language','Company size','Company type','Company type'])
```

# Filling Values with 0

```python

df4["Annual_Incentives"]=df4['Annual_Incentives'].fillna(0)
```

```python
df4.isnull().sum()
```

```python
#df4['Annual_Incentives']=df4["Annual_Incentives"].astype('int')
```

# Adding Custom Column Total_Salary

```python
df4['Total_Salary'] = df4['Salary_Without_Incentives'] + df4['Annual_Incentives']
```

```python
df4
```

# Replacing Values 

```python
replacement_mapping_dict = {
    "M": "Male",
    "F": "Female"
}
df4['Gender'] = df4['Gender'].replace(replacement_mapping_dict)
```
